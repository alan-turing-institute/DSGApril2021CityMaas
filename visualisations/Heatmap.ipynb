```python
import pandas as pd
import matplotlib as plt
```

```python
lon_acc_wheel = pd.read_csv('/shared/lon_wheel_acc.csv')
```

```python
#convert lon lat to float
lon_acc_wheel['lat'] = lon_acc_wheel['lat'].astype(float)
lon_acc_wheel['lon'] = lon_acc_wheel['lon'].astype(float)
```

```python
#convert to array
lon_acc_arr = lon_acc_wheel[['lat','lon']].values
```

```python
import folium
```

```python
lon_osm = folium.Map(location=[51.5074, 0.1278], zoom_start = 10, control_scale = True)
```

```python
from folium.plugins import HeatMap
```

```python
acc = HeatMap(lon_acc_arr).add_to(lon_osm)
acc.save('/shared/heatmap.html')
```
