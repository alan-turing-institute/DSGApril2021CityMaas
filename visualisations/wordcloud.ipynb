```python
import pandas as pd
#import wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_pois_wordcloud.csv')
df_accessible = pd.read_csv('accessible_pois_wordcloud.csv')
```

```python
_ = df_accessible['properties.other_tags']
_ = _.apply(lambda x: str(x).lower())
_ = _.apply(lambda x: x.replace('"','').replace('\'',''))

_ = _.apply(lambda x: x.split(','))
_ = _.apply(lambda x: [y[0:y.find('=>')] for y in x])

acc = []
for a in _:
    acc.extend(a)

_acc = " ".join(acc)
```

```python
wc = WordCloud(width = 800, height = 800, background_color='white', collocations=False).generate(_acc)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wc)
plt.savefig('wordcloud_accessible.png')
```

```python
_acc = []
for x in acc:
    _acc.append(" " + x[0:max(len(x),x.find(':'))])

_acc = " ".join(_acc)
```

```python
wordcloud = WordCloud(width = 800, height = 800, background_color='white', collocations=False).generate(_acc)

plt.figure(figsize = (10,10))
plt.imshow(wordcloud)

plt.savefig('wordcloud_accessible_merged.png')
```

```python
_acc = []
for x in acc:
    if len(x) > x.find(':'):
        _acc.append(" " + x[0:max(len(x),x.find(':'))])

_acc = " ".join(_acc)
```

```python
wordcloud = WordCloud(width = 800, height = 800, background_color='white', collocations=False).generate(_acc)

plt.figure(figsize = (10,10))
plt.imshow(wordcloud)

plt.savefig('wordcloud_accessible_merged_noemptytags.png')
```
