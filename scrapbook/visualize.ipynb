```python
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-colorblind')
```

```python
df = pd.read_pickle('/shared/accessible_pois.pkl')
df.loc[df['access']=='NONE'] = None
df
```

```python
fig,ax = plt.subplots(figsize=(8,6))
df['barrier'].value_counts(normalize=True).plot(kind='bar')
plt.savefig('dist_barrier.png')

fig,ax = plt.subplots(figsize=(8,6))
df['access'].value_counts(normalize=True).plot(kind='bar')
plt.savefig('dist_access.png')

fig,ax = plt.subplots(figsize=(8,6))
df['motor_vehicle'].value_counts(normalize=True).plot(kind='bar')
plt.savefig('dist_motor.png')

fig,ax = plt.subplots(figsize=(8,6))
df['amenity'].value_counts(normalize=True).plot(kind='bar')
plt.savefig('dist_amenity.png')
```
